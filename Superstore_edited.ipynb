```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'US Superstore data.csv')
```

```python
df.head()
```

```python
df=df.dropna()
```

```python
year=[]
for i in range(0,len(df)):
  year.append(df['Order Date'][i][6:])
```

```python
year
```

```python
for i in range(0,len(df)):
  year[i]=int(year[i])
```

```python
year
```

```python
df['Order_year']=year
```

```python
diffd=[]
diffm=[]
for i in range(0,len(df)):
    d1=int(df['Order Date'][i][:2])
    m1=int(df['Order Date'][i][3:5])
    d2=int(df['Ship Date'][i][:2])
    m2=int(df['Ship Date'][i][3:5])
    dres=d2-d1
    mres=m2-m1
    diffd.append(dres)
    diffm.append(mres)

```

```python
df['difference days']=diffd
df['difference months']=diffm
```

```python
profit_percentage=[]
for i in range(0,len(df)):
  profitp=df['Profit'][i]/df['Sales'][i]
  profit_percentage.append(profitp)
```

```python
for i in range(0,len(profit_percentage)):
  profit_percentage[i]=profit_percentage[i]*100
```

```python
loss_percentage=[]
for i in range(0,len(profit_percentage)):
  loss_percentage.append(1-profit_percentage[i])
```

```python
df['profit_percentage']=profit_percentage
df['loss_percentage']=loss_percentage
```

```python
quarter=[]
for i in range(0,len(df)):
  if(df['Order_month'][i]<=3):
    x='Q1'
    quarter.append(x)
  elif(df['Order_month'][i]>3 and df['Order_month'][i]<=6):
    x='Q2'
    quarter.append(x)
  elif(df['Order_month'][i]>6 and df['Order_month'][i]<=9):
    x='Q3'
    quarter.append(x)
  elif(df['Order_month'][i]>9):
    x='Q4'
    quarter.append(x)

```
